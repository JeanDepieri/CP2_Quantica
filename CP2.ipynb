```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# Criar circuito com 2 qubits e 2 bits clássicos
qc = QuantumCircuit(2, 2)

# Aplicar Hadamard e CNOT
qc.h(0)
qc.cx(0, 1)

# Medição dos qubits para os bits clássicos
qc.measure(0, 0)
qc.measure(1, 1)

# Visualizar circuito desenhado
qc.draw('mpl')
plt.show()

# Simulador
sim = AerSimulator()

# Compilar/transpilar para garantir que a medição será preservada
compiled = transpile(qc, sim)

# Executar simulação com 1000 shots
result = sim.run(compiled, shots=1000).result()

# Obter contagens
counts = result.get_counts()

# Mostrar contagens no console
print("Contagens:", counts)
%matplotlib inline
# Plotar histograma
plot_histogram(counts)
plt.title("Simulação com 1000 shots")
plt.show()
```
